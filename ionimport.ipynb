### Maricopa, Pinal, Pima

```python
import pandas as pd
import numpy as np
import re
from difflib import get_close_matches
```

```python
maricopa = pd.read_csv('auction_data/maricopa.csv')
pima = pd.read_csv('auction_data/pima.csv')
pinal = pd.read_csv('auction_data/pinal.csv')
```

#### Add unique ID's & Market info to each county csv

```python
#Maricopa

def format_apn(parcel):
    if pd.isna(parcel):
        return None
    
    parcel = parcel.strip()

    if len(parcel) == 8:
        return f"{parcel[:3]}-{parcel[3:5]}-{parcel[5:]}"
    elif len(parcel) == 9:
        return f"{parcel[:3]}-{parcel[3:5]}-{parcel[5:]}"
    else:
        return None

# Create APN column
maricopa["APN"] = maricopa["Parcel"].apply(format_apn)

#Create Market column
maricopa["Market"] = "Phoenix"
```

```python
maricopa.head()
```

```python
#Pima

pima["Parcel"] = pima["Parcel"].astype(str)

# Create APN column
pima["APN"] = (
    "PIMA_AZ."
    + pima["Parcel"].str[:3]
    + "-"
    + pima["Parcel"].str[3:5]
    + "-"
    + pima["Parcel"].str[5:]
)

#Create Market column
pima["Market"] = "Tucson"
```

```python
pima.head()
```

```python
#Pinal

pinal["Parcel"] = pinal["Parcel"].astype(str)

# Create APN column
pinal["APN"] = (
    "PINAL_AZ."
    + pinal["Parcel"].str[:3]
    + "-"
    + pinal["Parcel"].str[3:5]
    + "-"
    + pinal["Parcel"].str[5:]
)

#Create Market column
pinal["Market"] = "Phoenix"
```

#### Combine all data to one csv

```python
ionimport = pd.concat([maricopa, pima, pinal], ignore_index=True)

ionimport.to_csv("export_data/ionimport.csv", index=False)
```

#### Add columns for full address and properly formatted recording number

```python
#Full address
#Labeled as 'Last Name' for Salesforce Lead object purposes

ionimport["Last Name"] = (
    ionimport["PropertyAddress"].fillna("").astype(str)
    + ", "
    + ionimport["PropertyCity"].fillna("").astype(str)
    + ", "
    + ionimport["PropertyState"].fillna("").astype(str)
    + " "
    + ionimport["PropertyZip"].fillna("").astype(str)
)

ionimport["Last Name"] = ionimport["Last Name"].str.title()
```

#### Auction Trustee Lookup

```python
contacts = pd.read_csv('sf_data/trusteecontacts.csv')
```

```python
contacts.info()
```

```python
# Clean up column names if there are leading/trailing spaces
contacts.columns = contacts.columns.str.strip()
```

```python

def norm(s):
    if pd.isna(s):
        return ""
    s = str(s).upper()
    s = s.replace("\u00A0", " ")              # non-breaking spaces
    s = re.sub(r"[^\w\s]", " ", s)            # remove punctuation
    s = re.sub(r"\s+", " ", s).strip()        # collapse spaces
    return s

contacts["contact_name_norm"] = contacts["Contact: Contact Name"].map(norm)
contacts["assoc_norm"] = contacts["Associated Names"].map(norm)  # can be blank
```

```python
ionimport["trustee_norm"] = ionimport["Auction Trustee"].map(norm)
```

```python
exact_map = dict(zip(contacts["contact_name_norm"], contacts["Contact: Contact Name"]))
ionimport["Auction Trustee (Lookup)"] = ionimport["trustee_norm"].map(exact_map)
```

```python
def contains_in_contact_name(trustee_norm_value: str):
    if not trustee_norm_value:
        return None

    # find first contact where contact name contains trustee value
    hits = contacts.loc[
        contacts["contact_name_norm"].str.contains(re.escape(trustee_norm_value), na=False),
        "Contact: Contact Name"
    ]
    return hits.iloc[0] if len(hits) else None

mask = ionimport["Auction Trustee (Lookup)"].isna() & (ionimport["trustee_norm"] != "")
ionimport.loc[mask, "Auction Trustee (Lookup)"] = ionimport.loc[mask, "trustee_norm"].map(contains_in_contact_name)
```

```python
def contains_in_associated(trustee_norm_value: str):
    if not trustee_norm_value:
        return None

    hits = contacts.loc[
        contacts["assoc_norm"].str.contains(re.escape(trustee_norm_value), na=False),
        "Contact: Contact Name"
    ]
    return hits.iloc[0] if len(hits) else None

mask = ionimport["Auction Trustee (Lookup)"].isna() & (ionimport["trustee_norm"] != "")
ionimport.loc[mask, "Auction Trustee (Lookup)"] = ionimport.loc[mask, "trustee_norm"].map(contains_in_associated)
```

```python
unmatched = (
    ionimport.loc[ionimport["Auction Trustee (Lookup)"].isna(), "Auction Trustee"]
    .dropna()
    .drop_duplicates()
    .sort_values()
)
```
